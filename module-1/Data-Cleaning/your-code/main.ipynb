#### 3. Create a mysql engine to set the connection to the server. Check the connection details in [this link](https://relational.fit.cvut.cz/dataset/Stats).

```python
from sqlalchemy import create_engine
engine = create_engine('mysql+pymysql://guest:relational@relational.fit.cvut.cz/stats')
```

#### 4. Import the users table.

```python
data_users = pd.read_sql_query('SELECT * FROM stats.users', engine)
data_users.head()
```

#### 5. Rename Id column to userId.

```python
data_users = data_users.rename(columns={"Id": "userId"})
data_users.head()
```

#### 6. Import the posts table. 

```python
data_posts = pd.read_sql_query('SELECT * FROM stats.posts', engine)
```

```python
data_posts.head()
```

#### 7. Rename Id column to postId and OwnerUserId to userId.

```python
data_posts = data_posts.rename(columns={"Id": "postId", "OwnerUserId": "userId",})
data_posts.head()
```

#### 8. Define new dataframes for users and posts with the following selected columns:
**users columns**: userId, Reputation, Views, UpVotes, DownVotes  
**posts columns**: postId, Score, userID, ViewCount, CommentCount

```python
column_names =['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']
users_selected = data_users[column_names]
users_selected
```

```python
column_names2 =['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']
posts_selected = data_posts[column_names2]
posts_selected
```

#### 9. Merge the new dataframes you have created, of users and posts. 
You will need to make an inner [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
#I think this was an outer join --> make an inner join instead --> will only leave user ID
data_merge = users_selected.merge(posts_selected, left_on = 'userId', right_on = 'postId')
data_merge
```

```python
#inner join:
data_merge_inner = pd.merge(posts_selected, users_selected)
data_merge_inner

#data_merge_inner = pd.merge(posts_selected, users_selected, on = 'userId')
#data_merge_inner
```

#### 10. How many missing values do you have in your merged dataframe? On which columns?

```python
null_cols = data_merge_inner.isnull().sum()
null_cols[null_cols > 0]
```
