## Developing a Model

### Implementation: Define a Performance Metric
What is the performance meteric with which you will determine the performance of your model? Create a function that calculates this performance metric, and then returns the score. 

```python
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    # Your code here:
    r2 = r2_score(y_true, y_predict)
    return r2
```

### Implementation: Shuffle and Split Data
Split the data into the testing and training datasets. Shuffle the data as well to remove any bias in selecting the traing and test. 

```python
from sklearn.model_selection import train_test_split

X = df[df.columns.difference(["medv"])] #columns.difference gives us all columns despite 
                                                                #for the specified one(s)
y = df["medv"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True)
```

```python
display(X)
display(y)
```

----

## Analyzing Model Performance
Next, we are going to build a Random Forest Regressor, and test its performance with several different parameter settings.

### Learning Curves
Lets build the different models. Set the max_depth parameter to 2, 4, 6, 8, and 10 respectively. 

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
#calculating scores for train:
depths = [2,4,6,8,10]
score_train = []
score_test = []

for x in depths:
    rf = RandomForestRegressor(n_estimators=100, max_depth=x)
    rf.fit(X_train, y_train)
    y_train_pred = rf.predict(X_train)
    score_train_n = r2_score(y_train, y_train_pred)
    score_train.append(score_train_n)

score_train
```

```python
#calculating scores for test:
depths = [2,4,6,8,10]
score_test = []

for x in depths:
    rf = RandomForestRegressor(n_estimators=100, max_depth=x)
    rf.fit(X_test, y_test)
    y_test_pred = rf.predict(X_test)
    score_test_n = r2_score(y_test, y_test_pred)
    score_test.append(score_test_n)
    
score_test
```

Now, plot the score for each tree on the training set and on the testing set.

```python
#plotting
df = pd.DataFrame({'score_train': score_train,
                  'score_test': score_test}, index=depths)
ax = df.plot.bar()
```

What do these results tell you about the effect of the depth of the trees on the performance of the model?

```python
# higher depth leads to a hgiher accuracy
```

```python
#I would say the optimum lies where score-train and score_test have almost the same r^2 score. As while on the one 
#hand r^2 score of the training data must be slightly higher, while on the other hand it shouldn't be too high
#compared to the test_set due to overfitting
#thus, my answer would be 7

rf = RandomForestRegressor(n_estimators=100, max_depth=7)
rf.fit(X_train, y_train)
y_train_pred = rf.predict(X_train)
score_train_n = r2_score(y_train, y_train_pred)

rf = RandomForestRegressor(n_estimators=100, max_depth=7)
rf.fit(X_test, y_test)
y_test_pred = rf.predict(X_test)
score_test_n = r2_score(y_test, y_test_pred)
```
