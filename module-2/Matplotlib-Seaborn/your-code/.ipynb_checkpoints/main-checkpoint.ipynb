#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
from math import e
x = np.arange(0,100)
y = x**2
y2 = e**x

fig, [ax1, ax2] = plt.subplots(1,2)
ax1.plot(x,y)
ax1.set_title('x,y PLOT')   
ax1.set_xlabel('x label')     
ax1.set_ylabel('y label') 
#ax1.set_yscale("linear")
#ax1.set_xscale("linear")#only y_scale or also x_scale?

ax1.set_xlim([0,100])
ax1.set_ylim([0,10000])

ax2.plot(x,y2)
ax2.set_title('logarithmic scale (e^x)')
ax2.set_xlabel('x label')
ax2.set_ylabel('y2 label')
ax2.set_yscale("log")
ax2.set_xscale("log")#only y_scale or also x_scale?

fig.set_figheight(7)
fig.set_figwidth(15)
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
fitbit.columns
```

```python
listofplots = ["Minutes Lightly Active", "Minutes Very Active","Minutes Sedentary"]

x = fitbit["Minutes Lightly Active"]
y = fitbit["Steps"]
plt.scatter(x,y)
plt.xlabel('Minutes Lightly Active')
plt.ylabel('Steps')

x = fitbit["Minutes Very Active"]
y = fitbit["Steps"]
plt.scatter(fitbit["Minutes Very Active"],fitbit["Steps"])
plt.xlabel('Minutes Very Active')
plt.ylabel('Steps')


plt.scatter(fitbit["Minutes Sedentary"],fitbit["Steps"])
plt.xlabel('Minutes Sedentary')
plt.ylabel('Steps')
```

```python
#using a loop
listofplots = ["Minutes Lightly Active", "Minutes Very Active","Minutes Sedentary"]

plt.figure(figsize=(20,10))
for plot in listofplots:
    x = fitbit[plot]
    y = fitbit["Steps"]
    plt.scatter(x,y,label=plot)
    plt.xlabel('Minutes')
    plt.ylabel('Steps')
    plt.legend()
```
